```python
import pandas as pd
import mysql.connector
import os

# List of CSV files and their corresponding table names
csv_files = [
    ('customers.csv', 'customers'),
    ('orders.csv', 'orders'),
    ('sellers.csv', 'sellers'),
    ('products.csv', 'products'),
    ('geolocation.csv', 'geolocation'),
    ('payments.csv', 'payments'),
    ('order_items.csv','order_items')# Added payments.csv for specific handling
]

# Connect to the MySQL database
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='uzair@110ali7212',
    database='Ecommerce'
)
cursor = conn.cursor()

# Folder containing the CSV files
folder_path = 'D:\SQL Projects\ecommerce'

def get_sql_type(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return 'INT'
    elif pd.api.types.is_float_dtype(dtype):
        return 'FLOAT'
    elif pd.api.types.is_bool_dtype(dtype):
        return 'BOOLEAN'
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return 'DATETIME'
    else:
        return 'TEXT'

for csv_file, table_name in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Replace NaN with None to handle SQL NULL
    df = df.where(pd.notnull(df), None)
    
    # Debugging: Check for NaN values
    print(f"Processing {csv_file}")
    print(f"NaN values before replacement:\n{df.isnull().sum()}\n")

    # Clean column names
    df.columns = [col.replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

    # Generate the CREATE TABLE statement with appropriate data types
    columns = ', '.join([f'`{col}` {get_sql_type(df[col].dtype)}' for col in df.columns])
    create_table_query = f'CREATE TABLE IF NOT EXISTS `{table_name}` ({columns})'
    cursor.execute(create_table_query)

    # Insert DataFrame data into the MySQL table
    for _, row in df.iterrows():
        # Convert row to tuple and handle NaN/None explicitly
        values = tuple(None if pd.isna(x) else x for x in row)
        sql = f"INSERT INTO `{table_name}` ({', '.join(['`' + col + '`' for col in df.columns])}) VALUES ({', '.join(['%s'] * len(row))})"
        cursor.execute(sql, values)

    # Commit the transaction for the current CSV file
    conn.commit()

# Close the connection
conn.close()
```

```python
query = """ 
select 
year(order_purchase_timestamp) as order_year,count(order_id) as total_orders
from orders
group by order_year
order by total_orders desc 
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ["order_year","total_orders"])
data
```

```python
query = """ select 
distinct customer_state , count(customer_id)
from customers 
group by customer_state
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ["customer_state","customer_count"])
data
```

```python
query = """ select sum(case when payment_value >=1 then 1 else 0 end)/count(*)*100 as  paid_installments
from payments ;
"""
cur.execute(query)
data = cur.fetchall()

data = pd.DataFrame(data,columns = ["paid_installments"])
data
```

```python
query = """ 
select 
distinct c.customer_state , count(c.customer_id) as customer_count
from customers as c
join orders as o
on o.customer_id = c.customer_id 
where year(o.order_purchase_timestamp) = 2016
group by c.customer_state
order by customer_count desc; 
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data, columns = ["States","customer_count"])
data.head(5)
```

```python
query = """ 
select 
monthname(order_purchase_timestamp) as monthname, count(order_id) as order_count
from orders
where year(order_purchase_timestamp)= 2018
group by monthname(order_purchase_timestamp)
order by order_count desc;
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data, columns = ["Month","Total_orders"])
plt.figure(figsize =(8,3))
ax = sns.barplot(x = data['Month'], y= data['Total_orders'] , data=data)
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 45)
plt.title("orders per month in 2018")
plt.legend()
plt.grid()
plt.show()
```

```python
query = """ 
 with per_order as (
 select 
 o.order_id,o.customer_id,count(ot.order_id) as order_avg
 from orders as o
 join order_items as ot
 on ot.order_id = o.order_id
 group by o.order_id,o.customer_id)
 select cu.customer_city, round(avg(pr.order_avg),2) as avg_order
 from customers as cu
 join per_order as pr
 on cu.customer_id = pr.customer_id
 group by cu.customer_city
 order by cu.customer_city;
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data, columns = ["Customer_City","Average"])
data.head(5)
```

```python
query = """ 
select 
p.product_category,
round((sum(pmt.payment_value)/(select sum(payment_value) from payments))*100,2) as Revenue
from order_items as ot
join products as p
on ot.product_id = p.product_id
join payments as pmt
on pmt.order_id = ot.order_id
group by p.product_category
order by Revenue desc;
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ['Category','Revenue'])
data.head(5)
```

```python
query = """
select *,dense_rank() over(order by revenue desc) as rn from (select
ot.seller_id,round(sum(pmt.payment_value),2) as revenue
from order_items as ot
join payments as pmt
on ot.order_id = pmt.order_id
group by ot.seller_id) as a
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ['Top_Seller','Revenue','Rank'])
data.head(5)
```

```python
query = """
select  
_month,_year,payment, sum(payment) over w as commulative_sum
from
(select year(o.order_purchase_timestamp) as _year,
month(o.order_purchase_timestamp) as _month,
round(sum(p.payment_value),2) as payment
from orders as o
join payments as p
on o.order_id = p.order_id
group by _year,_month
order by _year) as a
group by _month,_year,payment
window w as (Rows between unbounded preceding and current row)
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ['Months','years','payments','Commulative_sum'])
data
```

```python
query = """
select  
customer_id,_year,payment, avg(payment) over w as rolling_avg
from
(select year(o.order_purchase_timestamp) as _year,
o.customer_id,
round(sum(p.payment_value),2) as payment
from orders as o
join payments as p
on o.order_id = p.order_id
group by _year,customer_id
order by _year) as a
group by _year,payment,customer_id
window w as (Rows between 3 preceding and current row);
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ['customer','years','payments','rolling_mean'])
data.head(10)
```

```python
query = """
select 
*,lag(present_growth) over (order by _year) as past_growth,
-- currentyear - previyear/previyear
((present_growth - lag(present_growth) over (order by _year))
/(lag(present_growth) over (order by _year)))*100 as year_by_year_growth
from (select year(o.order_purchase_timestamp) as _year,
round(sum(p.payment_value),2) as present_growth
from orders as o
join payments as p
on o.order_id = p.order_id
group by _year
order by _year) as a
"""
cur.execute(query)
data = cur.fetchall()
data = pd.DataFrame(data,columns = ['years','present_sales','past_sales','year_by_year_growth'])
data
```
